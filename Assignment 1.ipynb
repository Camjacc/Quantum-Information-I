```python
def X(): 
    op=sp.csr_matrix([[0,1],[1,0]])
    return op

def Z():
    op=sp.csr_matrix([[1,0],[0,-1]])
    return op
    
def Y():
    op=sp.csr_matrix([[0,-1j],[1j,0]])
    return op
```

```python
print(X())

print(Y())

print(Z())
```
